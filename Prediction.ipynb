In the rapidly growing used car market, determining the appropriate selling price for a vehicle is a complex task influenced by various factors such as the carâ€™s make, model, age, kilometers driven, condition, and other factors. Incorrect pricing can lead to financial losses for sellers and dissatisfaction for buyers. The goal of this project is to develop a predictive model that accurately estimates the selling price of used cars based on key attributes. By leveraging historical sales data and machine learning techniques, the model aims to assist sellers in setting competitive prices, thereby improving market efficiency and customer satisfaction.

```python
# List to store outliers for each numerical columns
outliers = []

# Iterate over each column and find outliers
for i in df_num.columns:
    Q1 = np.quantile(df_num[i],0.25) # First quantile
    Q3 = np.quantile(df_num[i],0.75) # Third quantile
    IQR = Q3-Q1 # Inter Quartile Range
    # Check condition for outliers
    outliers.append(df_num[i][ (df_num[i]<(Q1-(IQR*1.5))) | (df_num[i]>(Q3+(IQR*1.5))) ].unique())
    
# Display outlier values
pd.DataFrame({"Outliers":outliers},index=df_num.columns)
```

### Target encoding

```python
encoder ={}
for col in df_cat.columns:
    encoder[col] = {}
    for category in df_cat[col].unique():
        encoder[col][category] = df_num[df_cat[col]==category]["sale_price"].mean()
    df_cat[col] = df_cat[col].map(encoder[col])
```

```python
decoder = { col:{value:key for key,value in encoder[col].items()} for col in encoder.keys()}
```

### Predict train and test data to find metric

```python
X_train
```

```python
models = [linear,stochastic,ridge,lasso,elasticnet,rfregressor,gbregressor,abregressor,xgbregressor]
performance = pd.DataFrame({"Model":models})
performance["R-square (train)"] = performance["Model"].apply(lambda model : r2_score(y_train,model.predict(X_train)))
performance["R-square (test)"] = performance["Model"].apply(lambda model : r2_score(y_test,model.predict(X_test)))
performance["RMSE (train)"] = performance["Model"].apply(lambda model : mean_squared_error(y_train,model.predict(X_train))**0.5)
performance["RMSE (test)"] = performance["Model"].apply(lambda model : mean_squared_error(y_test,model.predict(X_test))**0.5)
performance["MAPE (train)"] = performance["Model"].apply(lambda model : mean_absolute_percentage_error(y_train,model.predict(X_train))**0.5)
performance["MAPE (test)"] = performance["Model"].apply(lambda model : mean_absolute_percentage_error(y_test,model.predict(X_test))**0.5)
```

```python
# Round of the metrics to 4 decimal places
performance["R-square (train)"] = performance["R-square (train)"].apply(lambda a : str(round(a,4)))
performance["R-square (test)"] = performance["R-square (test)"].apply(lambda a : str(round(a,4)))
performance["RMSE (train)"] = performance["RMSE (train)"].apply(lambda a : str(round(a,4)))
performance["RMSE (test)"] = performance["RMSE (test)"].apply(lambda a : str(round(a,4)))
performance["MAPE (train)"] = performance["MAPE (train)"].apply(lambda a : str(round(a,4)))
performance["MAPE (test)"] = performance["MAPE (test)"].apply(lambda a : str(round(a,4)))
```
